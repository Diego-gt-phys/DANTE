# MÃ©todo de Euler
## Ejemplo 1

$$ \frac{dy}{dx} = y \: ; \: y(0)=1 $$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# EcuaciÃ³n diferencial
def dydx(x, y):
    return y

# MÃ©todo de Euler
def euler(x0, y0, h, xf):
    n = (int)((xf - x0) / h)
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)
    x[0] = x0
    y[0] = y0
    for i in range(1, n + 1):
        y[i] = y[i - 1] + h * dydx(x[i - 1], y[i - 1])
        x[i] = x[i - 1] + h
    return x, y

# Condiciones iniciales
x0 = 0
y0 = 1
xf = 5

# Creo las funciones
x1, y1 = euler(x0, y0, 0.5, xf)
x2, y2 = euler(x0, y0, 0.1, xf)
x3, y3 = euler(x0, y0, 0.05, xf)
x4, y4 = euler(x0, y0, 0.01, xf)
x = np.linspace(x0, xf, 500)
y = np.exp(x)

# Generate a color palette
colors = sns.color_palette("Set1", 5)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = e^x$', color='black', linewidth=2)
plt.plot(x1, y1, label=r'$h=0.5$', color=colors[0], linewidth=1.5)
plt.plot(x2, y2, label=r'$h=0.1$', color=colors[1], linewidth=1.5)
plt.plot(x3, y3, label=r'$h=0.05$', color=colors[2], linewidth=1.5)
plt.plot(x4, y4, label=r'$h=0.01$', color=colors[3], linewidth=1.5)

# Add labels and title
plt.title(r'MÃ©todo de Euler', loc='left', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12, loc='center', fontweight='bold')
plt.ylabel('y', fontsize=12, loc='center', fontweight='bold')
plt.axhline(0, color='black', linewidth=1.0, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=1.0, linestyle='--')  # y-axis

# Set limits
plt.xlim(x[0], x[-1])
plt.ylim(0, y[-1])

# Add grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Configure ticks for all four sides
plt.tick_params(axis='both', which='major', direction='in', length=8, width=1.2, labelsize=12, top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=5, width=1.0, labelsize=10, top=True, right=True)
plt.minorticks_on()
# Customize tick spacing for more frequent ticks on x-axis

plt.gca().set_xticks(np.arange(0, xf, 0.5))  # Major x ticks every 0.5 units
#plt.gca().set_xticks(np.arange(0, xf, 0.1), minor=True)  # Minor x ticks every 0.1 units
plt.gca().set_yticks(np.arange(20, y[-1], 20))  # Major y ticks every 20 units
#plt.gca().set_yticks(np.arange(5, y[-1], 5), minor=True)  # Minor ticks every 5 units

# Set thicker axes for ROOT-style appearance
plt.gca().spines['top'].set_linewidth(1.2)
plt.gca().spines['right'].set_linewidth(1.2)
plt.gca().spines['bottom'].set_linewidth(1.2)
plt.gca().spines['left'].set_linewidth(1.2)

# Add a legend
plt.legend(fontsize=12, loc='upper left', frameon=False)

# Save the plot as a PDF
# plt.savefig("MÃ©todo de euler.pdf", format="pdf", bbox_inches="tight")  # 'bbox_inches="tight"' ensures no extra white space

# Show the plot
plt.tight_layout()
plt.show()
```

# MÃ©todo de Runge-Kutta

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# EcuaciÃ³n diferencial
def dydx(x, y):
    return y

# MÃ©todo de Runge-Kutta de cuarto orden
def runge_kutta(x0, y0, h, xf):
    n = int((xf - x0) / h)
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)
    x[0] = x0
    y[0] = y0
    for i in range(1, n + 1):
        k1 = h * dydx(x[i - 1], y[i - 1])
        k2 = h * dydx(x[i - 1] + h / 2, y[i - 1] + k1 / 2)
        k3 = h * dydx(x[i - 1] + h / 2, y[i - 1] + k2 / 2)
        k4 = h * dydx(x[i - 1] + h, y[i - 1] + k3)
        y[i] = y[i - 1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        x[i] = x[i - 1] + h
    return x, y

# Condiciones iniciales
x0 = 0
y0 = 1
xf = 5

# SoluciÃ³n con RK4 para diferentes tamaÃ±os de h
x1, y1 = runge_kutta(x0, y0, 1, xf)
x2, y2 = runge_kutta(x0, y0, 0.5, xf)
x3, y3 = runge_kutta(x0, y0, 0.1, xf)
x4, y4 = runge_kutta(x0, y0, 0.05, xf)
x = np.linspace(x0, xf, 500)
y = np.exp(x)

# Generate a color palette
colors = sns.color_palette("Set1", 5)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = e^x$', color='black', linewidth=2)
plt.plot(x1, y1, label=r'$h=1$', color=colors[0], linewidth=1.5)
plt.plot(x2, y2, label=r'$h=0.5$', color=colors[1], linewidth=1.5)
plt.plot(x3, y3, label=r'$h=0.1$', color=colors[2], linewidth=1.5)
plt.plot(x4, y4, label=r'$h=0.05$', color=colors[3], linewidth=1.5)

# Add labels and title
plt.title(r'MÃ©todo de Runge-Kutta', loc='left', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12, loc='center', fontweight='bold')
plt.ylabel('y', fontsize=12, loc='center', fontweight='bold')
plt.axhline(0, color='black', linewidth=1.0, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=1.0, linestyle='--')  # y-axis

# Set limits
plt.xlim(x[0], x[-1])
plt.ylim(0, y[-1])

# Add grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Configure ticks for all four sides
plt.tick_params(axis='both', which='major', direction='in', length=8, width=1.2, labelsize=12, top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=5, width=1.0, labelsize=10, top=True, right=True)
plt.minorticks_on()

# Customize tick spacing for more frequent ticks on x-axis
#plt.gca().set_xticks(np.arange(x0, xf + 0.1, 1))  # Major x ticks every 0.5 units
#plt.gca().set_yticks(np.arange(0, y[-1], 20))  # Major y ticks every 20 units

# Set thicker axes for ROOT-style appearance
plt.gca().spines['top'].set_linewidth(1.2)
plt.gca().spines['right'].set_linewidth(1.2)
plt.gca().spines['bottom'].set_linewidth(1.2)
plt.gca().spines['left'].set_linewidth(1.2)

# Add a legend
plt.legend(fontsize=12, loc='upper left', frameon=False)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Write the differential equation to solve
def dydx(y):
    return y

# Optimized Runge-Kutta 4th order method
def runge_kutta(x0, y0, h, xf):
    n = int((xf - x0) / h)
    
    # Initialize arrays for x and y
    x = np.linspace(x0, xf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    # Precompute h/2 to avoid redundant calculation
    h_half = h / 2
    dydx_vals = np.empty(4)  # Array to store dydx evaluations

    for i in range(1, n + 1):
        x_i_minus_1 = x[i - 1]
        y_i_minus_1 = y[i - 1]
        
        # Precompute dydx values and store them in dydx_vals
        dydx_vals[0] = dydx(y_i_minus_1)  # k1
        dydx_vals[1] = dydx(y_i_minus_1 + h_half * dydx_vals[0])  # k2
        dydx_vals[2] = dydx(y_i_minus_1 + h_half * dydx_vals[1])  # k3
        dydx_vals[3] = dydx(y_i_minus_1 + h * dydx_vals[2])  # k4
        
        # Compute y[i] using precomputed dydx values
        y[i] = y_i_minus_1 + (h / 6) * (dydx_vals[0] + 2 * dydx_vals[1] + 2 * dydx_vals[2] + dydx_vals[3])
    
    return x, y

# Write the initial conditions
x0 = 0
y0 = 1
h = 0.001
xf = 5

x,y = runge_kutta(x0, y0, h, xf)

x_teo = np.linspace(x0, xf, 500)
y_teo = np.exp(x_teo)

# Generate a color palette
colors = sns.color_palette("Set1", 5)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'Curva numÃ©rica', color=colors[0], linewidth=2) # Numerical solution
plt.plot(x_teo, y_teo, label=r'Curva analÃ­tica', color="black", linewidth=1, linestyle = '--') # Theoretical solution

# Add labels and title
plt.title(r'ResoluciÃ³n de Ecuaciones Diferenciales', loc='left', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12, loc='center', fontweight='bold')
plt.ylabel('y', fontsize=12, loc='center', fontweight='bold')
plt.axhline(0, color='black', linewidth=1.0, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=1.0, linestyle='--')  # y-axis

# Set limits
plt.xlim(x[0], x[-1])
#plt.ylim(y[0], 0)

# Add grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

# Configure ticks for all four sides
plt.tick_params(axis='both', which='major', direction='in', length=9, width=1.3, labelsize=12, top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=5, width=1.0, labelsize=10, top=True, right=True)
plt.minorticks_on()

# Customize tick spacing for more frequent ticks on x-axis
#plt.gca().set_xticks(np.arange(0, xf + 0.1, 1))  # Major x ticks every 0.5 units
#plt.gca().set_yticks(np.arange(-1, 0.1, 0.2))  # Major y ticks every 20 units

# Set thicker axes
plt.gca().spines['top'].set_linewidth(1.2)
plt.gca().spines['right'].set_linewidth(1.2)
plt.gca().spines['bottom'].set_linewidth(1.2)
plt.gca().spines['left'].set_linewidth(1.2)

# Add a legend
plt.legend(fontsize=12, loc='upper left', frameon=False)

# Save the plot as a PDF
#plt.savefig("Figura_1.pdf", format="pdf", bbox_inches="tight")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Write the differential equation to solve
def dydx(x,y):
    return np.exp(y)

# Non optimized Runge-Kutta method
def runge_kutta_old(x0, y0, h, xf):
    n = int((xf - x0) / h)
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)
    x[0] = x0
    y[0] = y0
    for i in range(1, n + 1):
        k1 = h * dydx(x[i - 1], y[i - 1])
        k2 = h * dydx(x[i - 1] + h / 2, y[i - 1] + k1 / 2)
        k3 = h * dydx(x[i - 1] + h / 2, y[i - 1] + k2 / 2)
        k4 = h * dydx(x[i - 1] + h, y[i - 1] + k3)
        y[i] = y[i - 1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        x[i] = x[i - 1] + h
    return x, y

# Optimized Runge-Kutta 4th order method
def runge_kutta(x0, y0, h, xf):
    n = int((xf - x0) / h)
    
    # Initialize arrays for x and y
    x = np.linspace(x0, xf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    # Precompute h/2 to avoid redundant calculation
    h_half = h / 2
    dydx_vals = np.empty(4)  # Array to store dydx evaluations

    for i in range(1, n + 1):
        x_i_minus_1 = x[i - 1] #DEBUG Esta variable se creo antes de que x fuera un np.linspace
        y_i_minus_1 = y[i - 1]
        
        # Precompute dydx values and store them in dydx_vals
        dydx_vals[0] = dydx(x_i_minus_1, y_i_minus_1)  # k1
        dydx_vals[1] = dydx(x_i_minus_1, y_i_minus_1 + h_half * dydx_vals[0])  # k2
        dydx_vals[2] = dydx(x_i_minus_1, y_i_minus_1 + h_half * dydx_vals[1])  # k3
        dydx_vals[3] = dydx(x_i_minus_1, y_i_minus_1 + h * dydx_vals[2])  # k4
        
        # Compute y[i] using precomputed dydx values
        y[i] = y_i_minus_1 + (h / 6) * (dydx_vals[0] + 2 * dydx_vals[1] + 2 * dydx_vals[2] + dydx_vals[3])
    
    return x, y

# Write the initial conditions
x0 = 0
y0 = -2
h = 0.8
xf = 7

# Solve the DE
x,y = runge_kutta(x0, y0, h, xf)
x_old, y_old = runge_kutta_old(x0, y0, h, xf)

# Calculate the exact solution
y_teo = np.linspace(-2, 1, 500)
x_teo = np.exp(2)-np.exp(-y_teo)


# Generate a color palette
colors = sns.color_palette("Set1", 5)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'MÃ©todo optimizado', color=colors[0], linewidth=2) # Numerical solution
plt.plot(x_old, y_old, label=r'MÃ©todo antiguo', color=colors[1], linewidth=2, linestyle = '-.')
plt.plot(x_teo, y_teo, label=r'SoluciÃ³n analÃ­tica', color="black", linewidth=1, linestyle = '--') # Theoretical solution

# Add labels and title
plt.title(r'ResoluciÃ³n de Ecuaciones Diferenciales', loc='left', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12, loc='center', fontweight='bold')
plt.ylabel('y', fontsize=12, loc='center', fontweight='bold')
plt.axhline(0, color='black', linewidth=1.0, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=1.0, linestyle='--')  # y-axis

# Set limits
plt.xlim(x[0], x[-1])
plt.ylim(-2, 1)

# Add grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

# Configure ticks for all four sides
plt.tick_params(axis='both', which='major', direction='in', length=9, width=1.3, labelsize=12, top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=5, width=1.0, labelsize=10, top=True, right=True)
plt.minorticks_on()

# Customize tick spacing for more frequent ticks on x-axis
#plt.gca().set_xticks(np.arange(0, xf + 0.1, 1))  # Major x ticks every 0.5 units
#plt.gca().set_yticks(np.arange(-1, 0.1, 0.2))  # Major y ticks every 20 units

# Set thicker axes
plt.gca().spines['top'].set_linewidth(1.2)
plt.gca().spines['right'].set_linewidth(1.2)
plt.gca().spines['bottom'].set_linewidth(1.2)
plt.gca().spines['left'].set_linewidth(1.2)

# Add a legend
plt.legend(fontsize=12, loc='upper left', frameon=False)

# Save the plot as a PDF
plt.savefig("Figura_1.pdf", format="pdf", bbox_inches="tight")

# Show the plot
plt.tight_layout()
plt.show()
```
